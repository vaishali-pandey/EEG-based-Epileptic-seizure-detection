```python
def convert_label(x):
    if (x > 1):
        return 0
    else:
        return x

# apply convert function
df["y"] = df['y'].apply(convert_label)
# eeg_dataframe["y"].value_counts()
```

```python

plt.style.use("dark_background")# to set the background colour 
target_column =df["y"]


ax = sns.countplot(x = target_column, label="Count")
ax.bar_label(ax.containers[0])# for displaying count at the top of the bar 

non_seizure, seizure = target_column.value_counts()
print('The number of trials for the non-seizure class is:', non_seizure)
print('The number of trials for the seizure class is:', seizure)
```

```python
y = df.iloc[:,179].values
seizure_index_list = []
normal_index_list = []
for i in range(len(y)):
    if (y[i] == 1):
        seizure_index_list.append(i)
    else:
        normal_index_list.append(i)
```

```python
# Convert dataframe to matrix. Row: data entries. Column: features
X = df.iloc[:,1:179].values
```
